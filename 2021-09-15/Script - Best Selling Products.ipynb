```python
import pandas as pd
import os 
```

```python
pwd = os.getcwd()
```

```python
orders = pd.read_csv(pwd + "\\orders.csv")
all_products = pd.read_csv(pwd + "\\all_products.csv")
```

```python
orders.head()
```

```python
all_products.head()
```

```python
orders_merged = pd.merge(left=orders, right=all_products[['product_id','product_name', 'distributor_id']], how='left', left_on='product_id', right_on='product_id')
orders_q4 = orders_merged[orders_merged['date'] > "2017-10-00"]
orders_q4.sort_values(by=('region'))
```

```python
orders_by_region = orders_q4.groupby(['region', 'product_name', 'distributor_id'])['no_units'].sum().reset_index().sort_values(by=['region', 'no_units'], ascending=[True, False])
orders_by_region 
```

```python
products_top_5 = orders_by_region.groupby('region').head(5)
products_top_5 = products_top_5.rename(columns={'no_units': 'total_units_sold'})
products_top_5
```
